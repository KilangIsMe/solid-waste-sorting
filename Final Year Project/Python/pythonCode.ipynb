```python
import serial
import math
import time
import cv2
from ultralytics import YOLO

arduino = serial.Serial('COM3', 9600) # COM is based on own setup
time.sleep(2)
arduino.reset_input_buffer()

#Robot DH Parameters (in cm)
d1 = 9.5
a0 = 1.0
a1 = 10.5
a2 = 14.7
a3 = 0
d4 = 0
d5 = 18

#Angle Limits
angle_limits = [
    (0, 180),  # Base
    (0, 165), # Shoulder
    (0, 150),  # Elbow
    (0, 180),  # Wrist Pitch
    (0, 180),  # Wrist Roll
    (25, 90)   # Gripper
]

def adjust_for_gripper(x, y, z, offset=d5):
    r = math.sqrt(x**2 + y**2)
    if r == 0:
        return x, y, z  # prevent divide by zero
    x_adj = x - (offset * x / r)
    y_adj = y - (offset * y / r)
    return x_adj, y_adj, z

#Inverse Kinematics Solver
def inverse_kinematics(x, y, z=0, theta6=90):
    z_eff = z - d1
    r = math.sqrt(x**2 + y**2)
    r_eff = math.sqrt(r**2 + z_eff**2)

    if r_eff > a1 + a2:
        print(f"[SKIPPED] Target ({x:.2f}, {y:.2f}, {z:.2f}) is out of reach!")
        return None

    theta1 = math.atan2(y, x)

    cos_theta3 = (r_eff**2 - a1**2 - a2**2) / (2 * a1 * a2)
    cos_theta3 = max(min(cos_theta3, 1), -1)
    theta3 = math.acos(cos_theta3)

    phi = math.atan2(z_eff, r)
    cos_psi = (a1**2 + r_eff**2 - a2**2) / (2 * a1 * r_eff)
    cos_psi = max(min(cos_psi, 1), -1)  # Clamp to avoid domain error
    psi = math.acos(cos_psi)
    theta2 = phi + psi

    theta4 = (theta3 - theta2) + math.radians(90)
    theta5 = math.radians(90)

    angles_rad = [theta1, theta2, theta3, theta4, theta5]
    angles_deg = [
        int(max(min(math.degrees(angle), lim[1]), lim[0]))
        for angle, lim in zip(angles_rad, angle_limits[:-1])
    ]
    angles_deg.append(int(max(min(theta6, angle_limits[5][1]), angle_limits[5][0])))
    
    return angles_deg

#Gripper Angle Mapper
def map_width_to_angle(width_px, min_cm=1, max_cm=5, min_angle=90, max_angle=30, px_per_cm=17.8):
    width_cm = width_px / px_per_cm
    width_cm = max(min_cm, min(max_cm, width_cm))  # Clamp
    return int(((width_cm - min_cm) / (max_cm - min_cm)) * (max_angle - min_angle) + min_angle)

def safe_move(from_angles, to_angles, lift_shoulder_angle=90, delay=2):
    # Step 1: Lift by moving shoulder to 90°
    lifted = from_angles[:]
    lifted[1] = lift_shoulder_angle  # Shoulder joint index is 1
    print(f"[SAFE MOVE] Lifting to: {lifted}")
    send_to_arduino(lifted)
    time.sleep(delay)

    # Step 2: Move to destination with same lifted shoulder
    mid = to_angles[:]
    mid[1] = lift_shoulder_angle
    print(f"[SAFE MOVE] Moving to mid-target: {mid}")
    send_to_arduino(mid)
    time.sleep(delay)

    # Step 3: Lower to actual target position
    print(f"[SAFE MOVE] Moving to final target: {to_angles}")
    send_to_arduino(to_angles)
    time.sleep(delay)


#Forward Kinematics Presets for Dropping
bin_angles = {
    "cardboard": [26, 165, 150, 97, 90, 25],
    "glass":     [38, 111, 150, 129, 90, 25],
    "metal":     [14, 89, 132, 132, 90, 25],
    "paper":     [35, 83, 103, 109, 90, 25],
    "plastic":   [29, 59, 94, 125, 90, 25],
    "default":   [60, 90, 90, 90, 90, 25]
}

#Neutral Home Position
home_angles = [90, 135, 135, 90, 90, 25]

# ----- Send to Arduino -----
def send_to_arduino(angles):
    data_string = ",".join(str(angle) for angle in angles) + "\n"
    arduino.write(data_string.encode())
    print(f"[SENT] {data_string.strip()}")

#Load custom trained YOLOv8 Model
model = YOLO("C:/Users/Kilang/runs/detect/train432/weights/best.pt")
cap = cv2.VideoCapture(0)

#Frame to Real World Conversion

def pixel_to_cm(x_pixel, y_pixel, px_per_cm=17.8, frame_w=640, frame_h=480):
    x_cm = ((x_pixel - frame_w / 2) / px_per_cm)
    y_cm = (frame_h - y_pixel) / px_per_cm                
    return x_cm, y_cm

#Detect Object and Execute Pick and Place
class_names = ['cardboard', 'glass', 'metal', 'paper', 'plastic']

while True:
    frame_read, frame = cap.read()
    if not frame_read:
        continue

    # Step 0: Apply roll correction
    h, w = frame.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, 2.83, 1.0)
    frame = cv2.warpAffine(frame, M, (w, h), flags=cv2.INTER_LINEAR)

    results = model(frame)
    detected_objects = []

    for r in results:
        for box in r.boxes:
            confidence = float(box.conf[0])
            if confidence >= 0.4:
                class_id = int(box.cls[0])
                obj_type = class_names[class_id]
                x1, y1, x2, y2 = box.xyxy[0]

                detected_objects.append({
                    'confidence': confidence,
                    'class_id': class_id,
                    'obj_type': obj_type,
                    'x1': x1,
                    'y1': y1,
                    'x2': x2,
                    'y2': y2
                })
            else:
                print(f"[SKIPPED] {obj_type} with confidence {confidence:.2f} is below threshold")

    if detected_objects:
        target = max(detected_objects, key=lambda x: x['confidence'])
        print(f"Processing highest confidence object: {target['obj_type']} ({target['confidence']:.2f})")

        # obtain coordinates
        mid_x = (target['x1'] + target['x2']) / 2
        mid_y = (target['y1'] + target['y2']) / 2

        # convert to real world coordinates
        x_cm, y_cm = pixel_to_cm(mid_x, mid_y)
        x_cm += 1
        y_cm += 22.7
        z_cm = 2

        # Draw box & label for visualization
        cv2.rectangle(frame, (int(target['x1']), int(target['y1'])), (int(target['x2']), int(target['y2'])), (0, 255, 0), 2)
        label = f"{target['obj_type']} {target['confidence']:.2f}"
        cv2.putText(frame, label, (int(target['x1']), int(target['y1']) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
        # timestamp = time.strftime("%Y%m%d-%H%M%S") 
        # cv2.imwrite(f"detected_{timestamp}.jpg", frame)
        # print(f"[SAVED] detected_{timestamp}.jpg")
        # uncomment to save the detected

        # set end effector width dynamically
        width_px = target['x2'] - target['x1']
        theta6 = map_width_to_angle(width_px)

        # gripper length offset
        x_adj, y_adj, z_adj = adjust_for_gripper(x_cm, y_cm, z_cm)
        print(f"[Adjusted] From (x={x_cm:.2f}, y={y_cm:.2f}) → Wrist (x={x_adj:.2f}, y={y_adj:.2f})")

        pickup_angles = inverse_kinematics(x_adj, y_adj, z_adj, theta6=theta6)
        if pickup_angles is None:
            continue
        print(f"[IK RESULT] x={x_adj:.2f} y={y_adj:.2f} z={z_adj} => Angles: {pickup_angles}")
        safe_move(home_angles, pickup_angles)

        final_pickup = pickup_angles[:]
        final_pickup[5] = max(min(pickup_angles[5] + 10, 80), 25)
        print(f"[SAFE MOVE] Clamping: {final_pickup}")
        send_to_arduino(final_pickup)
        time.sleep(2)

        # Obtain drop zone and move horizonotally to target
        drop_angles = bin_angles.get(target['obj_type'], bin_angles["default"])
        drop_angles[5] = final_pickup[5]
        print(f"[FK] Drop Angles for '{target['obj_type']}': {drop_angles}")
        safe_move(final_pickup, drop_angles)

        # Release
        drop_angles[5] = 25
        print(f"[SAFE MOVE] drop: {drop_angles}")
        send_to_arduino(drop_angles)
        time.sleep(2)

        # Step 3: Return to home position
        print(f"[HOME] Return to Home Angles: {home_angles}")
        send_to_arduino(home_angles)
        time.sleep(2)
        #safe_move(drop_angles, home_angles)

    # Show visual feedback
    cv2.imshow("Sorted View", frame)

    print("[READY] Press 'n' to process the next object, or 'q' to quit...")
    while True:
        key = cv2.waitKey(0) & 0xFF
        if key == ord('n'):
            break  # continue with next detection
        elif key == ord('q'):
            cap.release()
            cv2.destroyAllWindows()
            print("[EXIT] Program terminated by user.")
            break
        else:
            print("Press 'n' to continue, or 'q' to quit.")

    # use this instead of whil loop for auto sort
    # Press 'q' to quit loop
    #if cv2.waitKey(1) & 0xFF == ord('q'):
    #    break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
